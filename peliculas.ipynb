## CRÍTICAS DE PELÍCULAS

```python
# librerías

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Frame con las opinioones (de 0 a 5) obtenidas de las críticas de una película.

```python

df_votantes = {"Opinión (Xi)": [5,4,3,2,1,0], "Cantidad de votantes (Ni)": [40, 99,145,133,96,40]}
df_votantes = pd.DataFrame(df_votantes)
df_votantes
```

Representación gráfica de los datos

```python
plt.figure(figsize=(8, 6))

# Graficar los datos en un gráfico de barras
plt.bar(df_votantes['Opinión (Xi)'], df_votantes['Cantidad de votantes (Ni)'])

# Agregar una línea que una todas las barras
plt.plot(df_votantes['Opinión (Xi)'], df_votantes['Cantidad de votantes (Ni)'], color='red')

# Configurar etiquetas y título del gráfico
plt.xlabel('Opinión (Xi)')
plt.ylabel('Cantidad de votantes (Ni)')
plt.title('Gráfico de Barras con línea de unión')

# Mostrar el gráfico
plt.show()
```

Las opiniones corresponden a nuestros valores observados denominados Xi, y la cantidad de votantes se equipara a la cantidad de veces en que el valor observado ha sido elegido por los espectadores. Entonces hablamos de frecuencia de elección, que se denomina Ni.

A fin de calcular la media de esta serie de observaciones, para cada observación hay que realizar el producto de las opiniones por la cantidad de votantes:

```python
df_votantes["Productos"] = df_votantes["Opinión (Xi)"] * df_votantes["Cantidad de votantes (Ni)"]
df_votantes
```
